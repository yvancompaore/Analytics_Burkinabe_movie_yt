```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.action_chains import ActionChains
import time



# Setup Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Ensure GUI is off
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")


# Choose Chrome Browser
driver = webdriver.Chrome(service=webdriver_service, options=chrome_options)

# Set path to chromedriver as per your configuration
webdriver_service = Service(ChromeDriverManager().install())

# Ouvrir YouTube
driver.get('https://www.youtube.com/results?search_query=film+burkinabe')

# Fermer les pop-ups et bannières
try:
    consent_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="content"]/div[2]/div[6]/div[1]/ytd-button-renderer[2]/yt-button-shape/button/yt-touch-feedback-shape/div/div[2]'))
    )
    consent_button.click()
except Exception as e:
    print("Aucun pop-up de consentement trouvé ou erreur lors de la fermeture du pop-up :", e)

# Attendre que les vidéos soient chargées
videos = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.XPATH, '//*[@id="video-title"]'))
)

for video in videos:
    try:
        # Déplacer la souris sur la vidéo pour s'assurer qu'aucun élément ne bloque le clic
        actions = ActionChains(driver)
        actions.move_to_element(video).perform()

        # Utiliser JavaScript pour cliquer sur la vidéo
        driver.execute_script("arguments[0].click();", video)
        
        # Attendre que la durée soit chargée
        duration = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="movie_player"]/div[36]/div[2]/div[1]/div[1]/span[2]/span[3]'))
        )
        print(f"Durée de la vidéo : {duration.text}")
        
        # Revenir à la page des résultats
        driver.back()
        
        # Attendre que les vidéos soient de nouveau chargées
        videos = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, '//*[@id="video-title"]'))
        )
    except Exception as e:
        print(f"Une erreur est survenue : {e}")
        continue

# Fermer le driver
driver.quit()
```
