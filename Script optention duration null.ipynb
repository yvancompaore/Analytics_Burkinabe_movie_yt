```python
# Fonction pour convertir le temps en hh:mm:ss
def convert_to_hh_mm_ss(time_str):
    parts = list(map(int, re.split('[:]', time_str)))
    if len(parts) == 1:
        return f"00:00:{parts[0]:02d}"
    elif len(parts) == 2:
        return f"00:{parts[0]:02d}:{parts[1]:02d}"
    else:
        return f"{parts[0]:02d}:{parts[1]:02d}:{parts[2]:02d}"
```

```python

# Appliquer la fonction √† la colonne 'Temps'
df['duration'] = df['duration'].apply(convert_to_hh_mm_ss)
```

```python
# Fonction pour extraire et uniformiser les dates
def extract_date(text):
    # Regex pour trouver les dates au format "11 avr 2022" ou "11 avr. 2022"
    match = re.search(r'\d{1,2} \w{3,4}\.? \d{4}', text)
    if match:
        return match.group(0).replace('.', '')
    else:
        print(text)
        return None
```

```python
# Dictionnaire de traduction des mois fran√ßais en anglais
mois_translation = {
    'janv': 'Jan', 'f√©vr': 'Feb', 'mars': 'Mar', 'avr': 'Apr', 'mai': 'May', 'juin': 'Jun',
    'juil': 'Jul', 'ao√ªt': 'Aug', 'sept': 'Sep', 'oct': 'Oct', 'nov': 'Nov', 'd√©c': 'Dec'
}

# Fonction pour traduire les mois fran√ßais en anglais
def translate_month(date_str):
    if date_str is None:  # G√©rer les valeurs None
        return None
    for fr_month, en_month in mois_translation.items():
        if fr_month in date_str:
            date_str = date_str.replace(fr_month, en_month)
    return date_str

```

```python
# Fonction pour convertir les dur√©es en minutes
def duration_to_minutes(duration):
    if isinstance(duration, str):
        hours, minutes, seconds = map(int, duration.split(':'))
    elif isinstance(duration, time):
        hours = duration.hour
        minutes = duration.minute
        seconds = duration.second
    else:
        return None
    
    total_minutes = hours * 60 + minutes + seconds / 60
    return total_minutes
```

```python
# Convertir les dur√©es en objets datetime.time si elles sont des cha√Ænes de caract√®res
df['duration'] = pd.to_datetime(df['duration'], format='%H:%M:%S').dt.time

# Appliquer la fonction de conversion √† la colonne 'duration'
df['duration_minutes'] = df['duration'].apply(duration_to_minutes)
```

```python
# Fonction pour attribuer la cat√©gorie en fonction de la dur√©e en minutes
def categorize_duration(minutes):
    if minutes < 10:
        return 'Court-m√©trage'
    elif 10 <= minutes <= 30:
        return 'Moyen-m√©trage'
    else:
        return 'Long-m√©trage'
```
